```python
# Import dependices
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from api_keys import api_key
```

```python
# Create a call for the Census using our api_key

base_url = "https://api.census.gov/data/2016/acs/acs1?get="
variables = 'B01001_001E,B01002_001E,B01002_002E,B01002_003E'
district_base = "&for=congressional%20district:"
district = '*'
key = '&key='+api_key
url = base_url+variables+district_base+district+key
#print(url)
```

```python
# Create a function to return Pandas Data Frames from our pull requests

def getAPIData(pd_url):
    response = requests.get(pd_url)
    res = response.json()
    url_df = pd.DataFrame(res)
    return url_df
```

```python
# Try our function out to see if it works
# Get population data per state per distract from the Census

raw_dist_pop_df = getAPIData(url)
#raw_dist_pop_df.head()

# Change column names
raw_dist_pop_df.columns=['Population','Mean Age','Mean Age Male','Mean Age Female','State','Congressional Distract']
#raw_dist_pop_df.head()

# Drop first row
dist_pop_df = raw_dist_pop_df.drop(raw_dist_pop_df.index[0])
dist_pop_df.head()
```
