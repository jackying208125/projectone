```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```python
# read in cleaned dataset as dataframe in it's own jupyter notebook
data = pd.read_csv('Ownership_Data/Gun Ownership by Year.csv')
data.head()
```

```python
#Plot line graph for households who do have guns
data.plot.line(x='Year', y='Yes', c='red', fontsize=12,figsize=(12,8))
plt.title("Gun Ownership by Household", fontsize=32)
plt.xlabel('Year', fontsize=28)
plt.xticks(rotation=0, fontsize=15)
plt.legend(fontsize=15)

#Plot line graph for households who don't have guns
data.plot.line(x='Year', y='No', c='blue', fontsize=12,figsize=(12,8))
plt.title("Gun Ownership by Household", fontsize=32)
plt.xlabel('Year', fontsize=28)
plt.xticks(rotation=0, fontsize=15)
plt.legend(fontsize=15)

#Plot line graph for households who refused to answer if they have guns
data.plot.line(x='Year', y='Refused', c='yellow', fontsize=12,figsize=(12,8))
plt.title("Gun Ownership by Household", fontsize=32)
plt.xlabel('Year', fontsize=28)
plt.xticks(rotation=0, fontsize=15)
plt.legend(fontsize=15)

#Plot line graph for households who don't know if they have guns
data.plot.line(x='Year', y='Unknown', c='green', fontsize=12,figsize=(12,8))
plt.title("Gun Ownership by Household", fontsize=32)
plt.xlabel('Year', fontsize=28)
plt.xticks(rotation=0, fontsize=15)
plt.legend(fontsize=15)

plt.grid(True)

# Save the figure
#plt.savefig("")
```
